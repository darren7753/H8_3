```python
from sklearn.datasets import load_breast_cancer
```

```python
breast = load_breast_cancer()
```

```python
breast_data = breast.data
```

```python
breast_data.shape
```

```python
breast_labels = breast.target
```

```python
breast_labels.shape
```

```python
import numpy as np
```

```python
labels = np.reshape(breast_labels,(569,1))
```

```python
final_breast_data = np.concatenate([breast_data,labels],axis=1)
final_breast_data.shape
```

```python
import pandas as pd
```

```python
breast_dataset = pd.DataFrame(final_breast_data)
```

```python
features = breast.feature_names
```

```python
features_labels = np.append(features,"label")
```

```python
breast_dataset.columns = features_labels
```

```python
breast_dataset.head()
```

```python
breast_dataset["label"].replace(0,"Benign",inplace=True)
breast_dataset["label"].replace(1,"Malignant",inplace=True)
```

```python
breast_dataset.tail()
```

# Visualizing PCA

```python
from sklearn.preprocessing import StandardScaler

x = breast_dataset.loc[:,features].values
x = StandardScaler().fit_transform(x)
```

```python
x.shape
```

```python
pd.DataFrame(x).describe()
```

```python
feat_cols = ["feature" + str(i) for i in range(x.shape[1])]
normalised_breast = pd.DataFrame(x,columns=feat_cols)
normalised_breast.tail()
```

```python
from sklearn.decomposition import PCA

pca_breast = PCA(n_components=2)
principalComponents_breast = pca_breast.fit_transform(x)
```

```python
principal_breast_Df = pd.DataFrame(data=principalComponents_breast,columns=["PC1","PC2"])
principal_breast_Df.tail()
```

```python
pca_breast.explained_variance_ratio_
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt

%matplotlib inline

plt.figure()
plt.figure(figsize=(10,10))

plt.xlabel("PC1")
plt.ylabel("PC2")
targets = ["Benign","Malignant"]

colors = ["r","g"]

for target,color in zip(targets,colors):
    indicesToKeep = breast_dataset["label"] == target
    plt.scatter(principal_breast_Df.loc[indicesToKeep,"PC1"],
                principal_breast_Df.loc[indicesToKeep,"PC2"],
                c=color,
                s=50)
    
plt.legend(targets)
plt.show()
```

# K-Means Clustering

```python
from sklearn.cluster import KMeans
```

```python
X = -2 * np.random.rand(100,2)
X1 = 1 + 2 * np.random.rand(50,2)

X[50:100,:] = X1

plt.scatter(X[:,0],X[:,1],s=50,c="b")
plt.show()
```

```python
wcss = []

for i in range(1,11):
    kmeans = KMeans(n_clusters=i,init="k-means++",max_iter=300,n_init=10,random_state=0)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
    
plt.plot(range(1,11),wcss,marker="o")
plt.title("The Elbow Method")
plt.xlabel("Number of Clusters")
plt.ylabel("WCSS")
plt.show()
```

```python
Kmean = KMeans(n_clusters=2)
Kmean.fit(X)
```

```python
Kmean.cluster_centers_
```

```python
Kmean.cluster_centers_[0,:]
```

```python
plt.scatter(X[:,0],X[:,1],s=50,c="b")
plt.scatter(Kmean.cluster_centers_[0,:][0],Kmean.cluster_centers_[0,:][1],s=200,c="g",marker="s")
plt.scatter(Kmean.cluster_centers_[1,:][0],Kmean.cluster_centers_[1,:][1],s=200,c="g",marker="s")
plt.show()
```

```python
data = pd.read_csv("Wholesale customers data.csv")
data.head()
```

```python
data.describe()
```

```python
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)

pd.DataFrame(data_scaled).describe()
```

```python
kmeans = KMeans(n_clusters=2,init="k-means++")
kmeans.fit(data_scaled)
```

```python
kmeans.inertia_
```

```python
wcss = []

for i in range(1,20):
    kmeans = KMeans(n_clusters=i,init="k-means++",max_iter=300,n_init=10,random_state=0)
    kmeans.fit(data_scaled)
    wcss.append(kmeans.inertia_)
    
plt.plot(range(1,20),wcss,marker="o")
plt.title("The Elbow Method")
plt.xlabel("Number of Clusters")
plt.ylabel("WCSS")
plt.show()
```

```python
kmeans = KMeans(n_clusters=6,init="k-means++")
kmeans.fit(data_scaled)

pred = kmeans.predict(data_scaled)
```

```python
frame = pd.DataFrame(data_scaled)
frame["cluster"] = pred

frame["cluster"].value_counts()
```
