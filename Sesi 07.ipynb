```python
import numpy as np
import pandas as pd
```

```python
# https://github.com/ardhiraka/PFDS_sources/blob/master/Canada.xlsx?raw=true
```

```python
df_can = pd.read_excel("https://github.com/ardhiraka/PFDS_sources/blob/master/Canada.xlsx?raw=true",
                       sheet_name="Canada by Citizenship",
                       skiprows=range(20),
                       skipfooter=2)
df_can
```

```python
df_can = df_can.iloc[:195,:43]
df_can.shape
```

```python
df_can.info()
```

```python
df_can.columns.values
```

```python
df_can.index.values
```

```python
col = ["AREA","REG","DEV","Type","Coverage"]

df_can.drop(col,axis=1,inplace=True)
df_can.head(2)
```

```python
df_can.rename(columns={"OdName":"Country",
                       "AreaName":"Continent",
                       "RegName":"Region"},
              inplace=True)
df_can.columns
```

```python
df_can["Total"] = df_can.sum(axis=1)
df_can.head(2)
```

```python
df_can.isnull().sum()
```

```python
df_can.describe()
```

```python
df_can["Country"]
```

```python
df_can[["Country",1980,1981,1982,1983,1984,1985]]
```

```python
df_can.set_index("Country",inplace=True)
df_can.head(3)
```

```python
df_can.loc["Japan"]
```

```python
df_can.loc["Japan",[1980,1981,1982,1983]]
```

```python
df_can.columns = list(map(str,df_can.columns))
```

```python
df_can.loc["Japan",["1980","1981","1982","1983"]]
```

```python
years = list(map(str,range(1980,2014)))
years
```

# Area Plots

```python
df_can.sort_values(["Total"],ascending=False,axis=0,inplace=True)
df_top5 = df_can.head().copy()
df_top5 = df_top5[years].transpose()
df_top5
```

```python
df_top5.index = df_top5.index.map(int)

df_top5.plot(kind="area",
             stacked = False,
             figsize=(20,10))
plt.title("Imigration Trend of Top 5 Countries")
plt.ylabel("Number of Imigrants")
plt.xlabel("Years")
plt.show()
```

```python
df_top5.plot(kind="area",
             alpha = 0.25,
             stacked = False,
             figsize=(20,10))
plt.title("Imigration Trend of Top 5 Countries")
plt.ylabel("Number of Imigrants")
plt.xlabel("Years")
plt.show()
```

```python
ax = df_top5.plot(kind="area",
                  alpha=0.25,
                  figsize=(20,10))
ax.set_title("Imigration Trend of Top 5 Countries")
ax.set_ylabel("Number of Imigrants")
ax.set_xlabel("Years")
plt.show()
```

```python
df_least5 = df_can.tail(5).copy()
df_least5 = df_least5[years].transpose()

df_least5.plot(kind="area",
               alpha=0.45,
               stacked=True,
               figsize=(20,10))

ax.set_title("Imigration Trend of Least 5 Countries")
ax.set_ylabel("Number of Imigrants")
ax.set_xlabel("Years")

plt.show()
```

```python
ax = df_least5.plot(kind="area",
                    alpha=0.45,
                    figsize=(20,10))
ax.set_title("Imigration Trend of Least 5 Countries")
ax.set_ylabel("Number of Imigrants")
ax.set_xlabel("Years")
plt.show()
```

```python
df_can.loc[["Denmark","Norway","Sweden"],years]
```

```python
df_can.loc[["Denmark","Norway","Sweden"],years].plot.hist()
```

```python
df_t = df_can.loc[["Denmark","Norway","Sweden"],years].transpose()
df_t.head()
```

```python
df_t.plot(kind="hist",figsize=(10,6))

plt.title("Histogram of Imigration from Denmark, Norway, and Sweden from 1980 - 2013")

plt.ylabel("Number of Years")
plt.xlabel("Number of Imigrants")

plt.show()
```

```python
count,bin_edges = np.histogram(df_t,15)

df_t.plot(kind="hist",
          figsize=(10,6),
          bins=15,
          alpha=0.6,
          xticks=bin_edges,
          color=["coral","darkslateblue","mediumseagreen"])

plt.title("Histogram of Imigration from Denmark, Norway, and Sweden from 1980 - 2013")

plt.ylabel("Number of Years")
plt.xlabel("Number of Imigrants")

plt.show()
```

```python
count,bin_edges = np.histogram(df_t,15)

xmin = bin_edges[0] - 10
xmax = bin_edges[-1] + 10

df_t.plot(kind="hist",
          figsize=(10,6),
          bins=15,
          alpha=0.6,
          xticks=bin_edges,
          color=["coral","darkslateblue","mediumseagreen"],
          stacked=True,
          xlim=(xmin,xmax))

plt.title("Histogram of Imigration from Denmark, Norway, and Sweden from 1980 - 2013")

plt.ylabel("Number of Years")
plt.xlabel("Number of Imigrants")

plt.show()
```

# Pie Chart

```python
df_continents = df_can.groupby("Continent",axis=0).sum()
df_continents.head()
```

```python
df_continents["Total"].plot(kind="pie",
                            figsize=(5,6),
                            autopct="%.1f%%",
                            startangle=90,
                            shadow=True)

plt.show()
```

```python
colors_list = ["gold","yellowgreen","lightcoral","lightskyblue"]
explode_list = [0.1,0,0,0,0.1,0.1]

df_continents["Total"].plot(kind="pie",
                            figsize=(20,10),
                            autopct="%.1f%%",
                            startangle=90,
                            shadow=True,
                            labels=None,
                            pctdistance=1.12,
                            colors=colors_list,
                            explode=explode_list)

plt.legend(labels=df_continents.index,loc="upper left")

plt.show()
```
